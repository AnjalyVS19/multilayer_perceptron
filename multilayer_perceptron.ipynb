```python
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
y=[0,1,1,0]*1000
x=[[0,0],[0,1],[1,0],[1,1]]*1000
x
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,random_state=3)
```

```python
len(xtrain)
len(xtest)
```

```python
clf=MLPClassifier(hidden_layer_sizes=[2],activation='logistic',random_state=3,solver='sgd')
clf.fit(xtrain,ytrain)
```

```python
clf.n_layers_
```

```python
clf.hidden_layer_sizes
```

```python
clf.n_outputs_
```

```python
prediction=clf.predict(xtest)
print(clf.score(xtest,ytest))
```

```python
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
```

```python
digits=load_digits()
x=digits.data
y=digits.target
print(x.shape)
```

```python
print(y)
```

```python
print(x[0])
```

```python
print(x[0].reshape(8,8))
```

```python
import matplotlib.pyplot as plt
plt.imshow(x[0].reshape(8,8))
plt.show()
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,random_state=3)
clf=MLPClassifier(hidden_layer_sizes=[150,150],alpha=0.1,max_iter=500)
clf.fit(xtrain,ytrain)
```

```python
clf.n_layers_
```

```python
clf.n_outputs_
```

```python
predictions=clf.predict(xtest)
```

```python
predictions
```

```python
clf.score(xtest,ytest)
```

```python
a=clf.predict_proba(xtest)
a[0]
```
